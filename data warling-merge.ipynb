  Parameters
    ----------
    left : DataFrame or named Series

        Object to merge with.
    how : {'left', 'right', 'outer', 'inner', 'cross'}, default 'inner'
        Type of merge to be performed.

        * left: use only keys from left frame, similar to a SQL left outer join;
          preserve key order.
        * right: use only keys from right frame, similar to a SQL right outer join;
          preserve key order.
        * outer: use union of keys from both frames, similar to a SQL full outer
          join; sort keys lexicographically.
        * inner: use intersection of keys from both frames, similar to a SQL inner
          join; preserve the order of the left keys.
        * cross: creates the cartesian product from both frames, preserves the order
          of the left keys.

Merge 

1. one-to-one  : joining two DataFrame objects on their indexes which must contain unique value
2. many-to-one : Joining a unique index to one or more columns in a different DataFrame
3. many-to-many: joining columns on columns

```python
merge1=pd.DataFrame({'letter':['A','B','C','D','E'],
                     'number':[1,2,3,4,5]
                    })
merge2=pd.DataFrame({'letter':['F','B','G','D','A,'],
                     'number':[6,2,7,4,1]
                    })
```

Type of merge Opertation

1) inner join
2) left join / left outer join
3) right join / right outer join
4) outer join / full outer join
5) cross join 

```python
merge1=pd.DataFrame({'letter':['A','B','C','D','E'],
                     'number':[1,2,3,4,5]
                    })
merge2=pd.DataFrame({'letter':['F','B','G','D','A,'],
                     'number':[6,2,7,4,1]
                    })
```

```python
merge1.merge(merge2,how='inner', on='number')
```

```python
merge1.merge(merge2, how='outer', on='number')
```

Reshaping and pivoting

pivot() pivot_table()
stack() unstack()
melt()
crosstab()

```python
# pivot : Reshape data by turning unique values from  one column into multiple columns

data={'Date':['2023-01-01','2023-01-01','2023-01-02','2023-01-02'],
      'city':['new york','los angeles','new york','los angeles'],
      'Temperature':[20,75,32,77] }
df=pd.DataFrame(data)
```

```python
df
```

```python
df_pivot=df.pivot(index='Date',columns='city',values='Temperature')
df_pivot
```

```python
pivot_table=df.pivot_table(index='Date',columns='city',values='Temperature',aggfunc='mean')
pivot_table
```

```python
stack_df=df.stack()
stack_df
```

```python
unstack_df=stack_df.unstack()
unstack_df
```

```python
melt_df=df.melt(id_vars=['Date','city'],value_vars=['Temperature'])
melt_df
```

```python
crosstab_df=pd.crosstab(df['Date'],df['city'])
crosstab_df
```

Data Aggregation

Data aggregation in pandas is the process of summarizing or transforming data to provide insights.
involves grouping the data one or more columns and the appluing aggregation function
1. sum()
2. mean()
3. max()
4. min()
5. count()
6. size()
7. std()
8. var()

```python
agg={'category':['A','B','A','B','A'],
    'value':[10,20,30,40,50]
    }
agg_df=pd.DataFrame(agg)
```

```python
agg_df
```

```python
groped=agg_df.groupby('category')['value'].sum()
groped
```

```python
agg1={'category':['A','B','A','B','A'],
    'value1':[10,20,30,40,50],
     'value2':[5,15,25,35,45]
    }
agg_df1=pd.DataFrame(agg1)
```

```python
agg_df1
```

```python
result=agg_df1.groupby('category')['value1'].agg({ 'sum','mean','max'})
result
```

```python
result1=agg_df1.groupby('category').agg({ 'value1':'std',
                                        # 'value1':'mean'
                                         #'value1':'max',
                                        #'value2':'sum',
                                         #'value2':'mean','value2':'max'
                                        })
```
